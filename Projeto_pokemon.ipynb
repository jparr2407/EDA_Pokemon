# Análise Exploratória

## Tipos de Pokemons mais frequentes

```python
# Substituir "-" por "não há" na coluna "Type 2"
df['Type 2'] = df['Type 2'].replace('-', 'Não há')

# Agora, vamos contar a frequência dos tipos nas colunas "Type 1" e "Type 2"
top_type1 = df['Type 1'].value_counts().head(5)
top_type2 = df['Type 2'].value_counts().head(5)

# Criar uma figura com dois subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Gráfico para os top 5 tipos em "Type 1"
ax1.bar(top_type1.index, top_type1.values, color='skyblue')
ax1.set_title('Tipos Primários mais Frequentes')
ax1.set_xlabel('Tipo 1')
ax1.set_ylabel('Contagem')

# Gráfico para os top 5 tipos em "Type 2"
ax2.bar(top_type2.index, top_type2.values, color='lightcoral')
ax2.set_title('Tipos Secundários mais Frequentes')
ax2.set_xlabel('Tipo 2')
ax2.set_ylabel('Contagem')

# Ajustar o layout para evitar sobreposição de rótulos
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```

**Frequência dos Tipos Primários (Type 1):** O gráfico da esquerda mostra os tipos primários mais frequentes nos Pokémons. Isso ajuda a identificar os tipos de Pokémon mais comuns entre eles.

**Frequência dos Tipos Secundários (Type 2):** O gráfico da direita apresenta os tipos secundários mais frequentes nos Pokémons. Isso pode ser útil para entender quais tipos secundários são mais comuns entre os Pokémon com duas tipagens. Nesse gráfico há maior aparição dos dados definidos como "Não há", essa representação é por que alguns Pokemons não possuírem 2 tipos, então quando está escrito "Não há", o pokemon em questão só possui 1 tipo.

Aqui podemos observar que o tipo primário mais comum é água, e a grande maioria dos pokemons não possuem o tipo secundário.

## Pokemons com maior quantidade de evoluções

```python
# Contar as frequências dos nomes base (Base Name)
top_base_names = df['Base Name'].value_counts().head(10)

# Criar um gráfico de barras usando seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=top_base_names.index, y=top_base_names.values, palette="viridis")
plt.title('Pokemons com mais evoluções"')
plt.xlabel('Nome Base')
plt.ylabel('Frequência')
plt.xticks(rotation=45)  # Rotacionar rótulos do eixo x para melhor legibilidade
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```

**Análise de Popularidade:** A contagem dos nomes base pode ser usada como um indicador de popularidade de determinadas evoluções ou formas de Pokémon.

Aqui podemos observar que o pokemon que possui a maior quantidade de evoluções e transformações é o Rotom.

## Histograma dos atributos dos pokemons

```python
# Selecionar as colunas de interesse
colunas_atributos = ["BST", "HP", "Attack", "Defense", "Sp. Attack", "Sp. Defense", "Speed"]

# Criar subplots de histogramas para cada coluna
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
fig.suptitle('Histogramas dos Atributos dos Pokemons', fontsize=16)

# Iterar sobre as colunas e criar os histogramas
for i, column in enumerate(colunas_atributos):
    row = i // 4
    col = i % 4
    sns.histplot(df[column], kde=True, ax=axes[row, col], color='skyblue')
    axes[row, col].set_title(column, fontsize=12)
    axes[row, col].set_xlabel(None)

# Ajustar o layout
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Mostrar o gráfico
plt.show()
```

**Distribuição dos Atributos:** Os histogramas mostram a distribuição dos valores dos atributos dos Pokémon, incluindo BST (Total de Base Stats), HP, Attack, Defense, Sp. Attack, Sp. Defense e Speed.

**Identificação de Padrões:** Você pode identificar padrões nas distribuições, como a forma da curva (normal, assimétrica, bimodal) e a faixa de valores em que a maioria dos Pokémon se encontra.

**Detecção de Outliers:** Os histogramas podem ajudar a identificar a presença de outliers (valores extremos) nos atributos.


Aqui é possível realizar uma análise visual dos atributos dos Pokémon, permitindo a identificação de padrões e insights sobre como esses atributos estão distribuídos. Por exemplo:

*   BST mais comum: entre 400 e 600
*   HP mais comum: entre 50 e 100
*   Attack mais comum: entre 50 e 100







## Pokemons com maior diferença de poder entre suas evoluções

```python
# Calcular a diferença no BST entre evoluções
df['BST Difference'] = df.groupby('Base Name')['BST'].diff().fillna(0)

# Selecionar os 10 maiores aumentos no BST
top_10 = df.nlargest(10, 'BST Difference')

# Criar um gráfico de barras para mostrar os top 10 aumentos no BST
plt.figure(figsize=(10, 6))
sns.barplot(data=top_10, x='Base Name', y='BST Difference', palette='viridis')
plt.title('Pokemons com maior diferença de poder entre suas evoluções')
plt.xlabel('Nome da Forma Base')
plt.ylabel('Aumento no BST')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

**Aumento no BST:** O código calcula a diferença no BST (Total de Base Stats) entre as diferentes evoluções de Pokémon. Isso permite identificar o aumento de poder ao evoluir.

**Identificação de Evoluções Poderosas:** O gráfico mostra quais evoluções têm os maiores aumentos no BST, o que pode indicar quais Pokémon se tornam mais poderosos quando evoluem.

**Comparações entre Evoluções:** É possível comparar as diferenças de poder entre as evoluções de diferentes Pokémon e entender quais evoluções proporcionam os maiores ganhos em BST.

Este código fornece uma análise dos aumentos de poder (BST) entre as evoluções dos Pokémon, destacando os 10 maiores aumentos e facilitando a identificação das evoluções mais poderosas.

Aqui podemos analisar que tem 2 pokemons que possuem evoluções fora da curva, que são muito superiores aos outros, o Wishiwashi e o Eternatus.

## Maiores atributos por tipo de pokemon

**Identificação dos pontos fortes:** Nessa análise é possível identificar quais tipos de pokemons que possuem os melhores atributos base.

**Comparação de atributos:** É possível comparar o que os tipos de pokemons tem de vantagem sobre os outros tipos.

### Speed (Velocidade)

```python
# Calcular a média de Speed por tipo
avg_speed_by_type = df.groupby('Type 1')['Speed'].mean().reset_index()

# Ordenar os tipos por média de Speed
avg_speed_by_type = avg_speed_by_type.sort_values(by='Speed', ascending=False)

# Criar um gráfico de barras para mostrar os tipos de Pokémon com maior média de Speed
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_speed_by_type, x='Speed', y='Type 1', palette='viridis')
plt.title('Tipos de Pokémon com Maior Média de Speed')
plt.xlabel('Média de Speed')
plt.ylabel('Tipo de Pokémon')

plt.tight_layout()
plt.show()
```

Aqui é possível identificar que os 3 tipos de pokemons que possuem:

*   **Maior Velocidade:** Elétrico, Voador e Dragão.
*   **Menor Velocidade:** Grama, Pedra e Aço.

### HP (Vida)

```python
# Calcular a média de HP por tipo
avg_hp_by_type = df.groupby('Type 1')['HP'].mean().reset_index()

# Ordenar os tipos por média de HP
avg_hp_by_type = avg_hp_by_type.sort_values(by='HP', ascending=False)

# Criar um gráfico de barras para mostrar os tipos de Pokémon com maior média de HP
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_hp_by_type, x='HP', y='Type 1', palette='viridis')
plt.title('Tipos de Pokémon com Maior Média de HP')
plt.xlabel('Média de HP')
plt.ylabel('Tipo de Pokémon')

plt.tight_layout()
plt.show()
```

Aqui é possível identificar que os 3 tipos de pokemons que possuem:

*   **Maior HP(vida):** Dragão, Normal e Lutador.
*   **Menor HP(vida):** Fantasma, Elétrico e Inseto.

## Correlação entre atributos de pokemons

```python
# Selecionar as colunas de interesse
cols_status = ["HP", "Attack", "Defense", "Sp. Attack", "Sp. Defense", "Speed"]
correlation_matrix = df[cols_status].corr()

# Criar um gráfico de correlação usando seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlação entre atributos")
plt.show()
```

**Matriz de Correlação:** O código calcula a matriz de correlação entre os status dos Pokémon, incluindo HP, Attack, Defense, Sp. Attack, Sp. Defense e Speed.

**Visualização de Correlações:** O gráfico de calor (heatmap) mostra visualmente as correlações entre os pares de atributos. Cores mais escuras indicam correlações mais fortes (positivas ou negativas), enquanto cores mais claras indicam correlações mais fracas ou próximas de zero.

**Importância para Estratégia:** Para jogadores de Pokémon, essa matriz de correlação pode ser útil na seleção de Pokémon com base em estratégias de batalha. Por exemplo, se Attack e Speed tiverem uma correlação positiva, pode ser vantajoso escolher Pokémon que tenham ambos os atributos elevados.



*   Os atributos Defesa e Velocidade possuem uma correlação negativa forte.
*   Os atributos Defesa especial e Defesa possuem uma correlação positiva moderada.
*   Os atributos Ataque especial e Ataque possuem uma correlação positiva moderada.
